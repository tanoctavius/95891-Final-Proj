```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stopwords = set(stopwords.words("english"))
```

```python
with open("/content/drive/MyDrive/AI final/cleaned data/captions_10k.json", "r") as f:
  captions = json.load(f)
```

```python
def clean_captions(s):
  sub_strs = s.lower().split()
  kept_words = [obj for obj in sub_strs if obj not in stopwords]

  res_str = " ".join(kept_words)

  doc = nlp(res_str)

  keywords = [tok.text for tok in doc if tok.pos_ in ["NOUN", "ADJ"]]

  return " ".join(keywords)
```

```python
cleaned_captions = {}

for key, caption in captions.items():
  cleaned_captions[key] = clean_captions(caption)
```

```python
with open("/content/drive/MyDrive/AI final/cleaned data/cleaned_captions.json", "w") as f:
  json.dump(cleaned_captions, f, indent=2)
```
