```python
from transformers import VisionEncoderDecoderModel, ViTImageProcessor, AutoTokenizer
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from PIL import Image
from PIL import ImageFile

import json
import os
import random
from tqdm import tqdm
```

```python
with open("/content/drive/MyDrive/AI final/cleaned data/captions_10k.json", "r") as f:
  captions = json.load(f)

# Randomize
keys = list(captions.keys())
random.shuffle(keys)

split_index = int(len(keys) * 0.8)

train_keys = keys[:split_index]
test_keys = keys[split_index:]
```

```python
with open("/content/drive/MyDrive/AI final/cleaned data/train.txt", "w") as f:
  for key in train_keys:
    f.write(key + "\n")
```

```python
with open("/content/drive/MyDrive/AI final/cleaned data/test.txt", "w") as f:
  for key in test_keys:
    f.write(key + "\n")
```
